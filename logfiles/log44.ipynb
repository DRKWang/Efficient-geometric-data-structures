### Generating  projecting directions based on inclination of cones of a given fan

In order to make the directions more seperatable with respect to cones of fan, we could generate certain directions based on the inclination of cones of a given fan instead of randomly.

Suppose the dimension of universal space is $n$. Concretely, 

1. We will compute the compute the inclination first. We measure this inclination of fan with "average point" $q$. The average point is defined by $q = \sum p_i$, $\{p_i\}$ are the points along with the generating rays of the fan and whose 1-norm are exactly 1.

2. Then, choose a set of orthogonal basis $\{v_j\}$ coming from the complementary space of the space generated by $q$, i.e. $\forall i$, $v_j  \perp q$. Note that, since the average point $q$ could be zero vector, in that case, we will choose $N(1,0,...0)$, $N(0,1,..., 0)$,..., $N(0,0,...,1)$ as the basis. The chosen set of orthogonal basis  $\{v_j\}$ could be good candidates for seperating the cones of fan. One way of generating the orthogonal basis $\{v_j\}$ will be using [Gramâ€“Schmidt process](https://en.wikipedia.org/wiki/Gram%E2%80%93Schmidt_process).

3. After generating orthogonal basis $\{v_j\}$, those basis will stored in the class `RationalPolyhedralFan` as atributes of the class. and those atributes will be invoked for generating new boxes for `_contains()`.


`allow_arragement` means allowing generate the fan based on the cones of input and including their intersection.

# Rtree Implementation

In general, there are many ways to implement Rtree including its variants, R+tree, R\*tree, ...., because is no significate difference between them in terms of performance, it is better to prefer simple and easy way to implement it firstly, then extend it gradually. 

(There are two basic measurement for the performance of rtree, **Coverage** and **Overlap**. Coverage is the entire area to cover all related rectangles. Overlap is the entire area which is contained in two or more nodes. Minimal coverage reduces the amount of "dead space" (empty area) which is covered by the nodes of the R-tree. Minimal overlap reduces the set of search paths to the leaves (even more critical for the access time than minimal coverage). Efficient search requires minimal coverage and overlap.)

We may classify implementation of the rtree into 2 classes, class `rtree` and class `node`.

### class `rtree`

##### attributes

- An array, which is used for storing all the input data for this `rtree`. This array is necessary, since it allows us to duplicate multiple boxes pointing to the same entry in rtree data structure.

- An rtree, a recursive tree data structure. Each node of it is made of left child node, right child node, and its data. The data is a tuple, $(((a_1,b_1),(a_2,b_2),...(a_n,b_n)), idx)$, where `idx` is the index of data in the array.

        Currently, my imagination is to build a complete binary tree for rtree and only keep each node identified to a single box. Thus, in that case, the rtree data structure can be implemented with an array. In detail, it will be as follows. .


- An integer for the length of the array for nodes.

- dim

We may a few confusing and similar definitions here. 

- The entry refers to the information from input, including (id, box, other information).
- The box refers to the coordiniates of the box.
- The data is in terms of node, referring to the comibination of box and the idx (only the leaf node can have `idx`, otherwise, the data of non-leaf nodes are just pure box without `idx`.
- The node is in terms of rtree, referring to the an atom for rtree, including a data the children node pointers.
