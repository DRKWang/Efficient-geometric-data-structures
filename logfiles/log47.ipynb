### Polyhedron fan

1. For the questions in fan, in a $d$ dimensional space, if we use a set of hyperplanes to intersect the cones, then their intersections will be $d-1$ geometric objects, and we hope to choose a proper direction and project them onto this 1 dimensional direction, which seems to be too much work.

2. We may use a different measurement, the angle between cones with a specific ray, instead of the linear projection.

To be specific, we can measure both the maximal angle distance and the minimal angle distance for a pair of cone and ray, then use this as the inputs x_max and x_min for a dimension of box in rtree.

It is easy to operation in 2 dimension, since we can use the generating rays of a cone to find the maximal angle and the minimal angle. But for dimension 3, we have to use kind of optimization to compute those two angles. But since we only focus on the overestimation, we may use some different way to do so. For example, we can think of using the spinning cone of a target cone, which is the circumscribed cone of the target cone, as the cone for measurement.
