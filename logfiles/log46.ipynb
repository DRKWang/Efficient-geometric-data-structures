##### compatibility of $\bot$

Following the previous discussion, 

we use $\bot$ to indicate the status of un-intersection of a ray and a projecting line/plane/hyperplane.

To make it well-defined, we will discuss a few observations and make a classifaction for cones.

Ex1,



Supposing that there is a cone, the red one, with a projecting line, the blue one, on a plane.




We hope to get the partial projecting result of this cone onto this line and try to compatible with real number system. Ideally, the result of the red cone should be an interval with $-\infty$, similar to the green cone. However, we can not make it. This is because intersections of generating rays and lines will be computed individually, without considering the mutual relationships. Thus, we will not know that the interval should be which one, $(-\infty, a)$ or $(a, +\infty)$. Thus, the best result is $\{\bot, a\}$.

Similarly, we will get the same result, $\{\bot, a\}$, from the orange cone. Ideally, we hope to let the result of red cone be $(-\infty, a]$ and result of orange cone be $[a, +\infty)$. But, their results are both $\{\bot, a\}$. Thus, to make it compatible with real number system and overestimating, we have to make $\{\bot, a\}$ replaced by $(-\infty, +\infty)$.

Ex2,

Apart from the following case, we may have another case, where the result of the cone only has $\bot$. For example, considering the purple cone. The result of this cone is $\{\bot,\bot\}$, which means we are not able to see this cone based on this projecting line. Based on that, we can conclude that this cone never has an intersection with a cone whose result is an limited interval, like the green cone. And it is possible that this cone has an intersection with a cone whose result including an $\bot$, which will be replaced by $(-\infty, +\infty)$. To follow the above discussion,  we could make $\{\bot,\bot\}$ replaced by $[A, +\infty)$, where $A$ should an extremely large number, in case that the limited interval has an intersection with $[A, +\infty)$.

Above all, we may have 3 types of result of cones.

1. all generating rays of the cone has intersections with the projecting line, -> a limited interval, $[a,b]$

2. partial generating rays of the cone has intersections with the projecting line, -> $(-\infty, +\infty)$

3. none of generating rays of the cone has intersections with the projecting line, -> $[A, +\infty)$

##### compress an unlimited region into an limited region.

Since the above method was based on locality, the non-importance will be higher as the reuslt goes infinity. Due to limited storage of modern computer, we may set some "focus region" to reduce the complexity. For example, we may set interval $[-2^{64}, 2^{64}]$ as the focus region and any interval that surpassing this region will be compressed into the corrosponding endpoints. For example, supposing the interval is $[100, 2^{70}]$, then we will let it be $[100, 2^{64}]$, and supposing the interval is $(-\infty, +\infty)$, then we will let it be $[-2^{64}, 2^{64}]$. Note that it is valid since we still keep it be overestimating. 

As a result, we can make the result of the 3rd type of cones be an endpoint, like $[2^{64}, 2^{64}]$.
